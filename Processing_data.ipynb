```
train_path = '/content/drive/My Drive/PROJECT 2/Train_Full/'
test_path = '/content/drive/My Drive/PROJECT 2/Test_Full/'


labels = os.listdir(train_path)
with open('/content/drive/My Drive/PROJECT 2/DATA/vietnamese-stopwords-dash.txt','r') as f:
    stop_words = f.read().splitlines()

def load_data(path):
  data = []
  label = []
  for index in range(10):
    dir_path = path + labels[index] + '/'
    for document in os.listdir(dir_path):
      with open(dir_path+document, encoding='utf-16') as f:
        doc = f.readlines()
        doc = ' '.join(doc)        
        doc = ViTokenizer.tokenize(doc)
        doc = doc.lower() 
        doc = gensim.utils.simple_preprocess(doc)
        doc = ' '.join(doc)
        tokens = doc.split() 
        tokens = [word for word in tokens if word not in stop_words]
        tokens = ' '.join(tokens)
        data.append(tokens)
        label.append(index)
  return data, label
```

```
x_train, y_train = load_data(train_path)

import pickle

pickle.dump(x_train, open('/content/drive/My Drive/PROJECT 2/DATA/x_train.pkl', 'wb'))
pickle.dump(y_train, open('/content/drive/My Drive/PROJECT 2/DATA/y_train.pkl', 'wb'))
```
